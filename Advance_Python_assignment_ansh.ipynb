```python
#ques 4
import random
import string
import os

def random_string(length=100):
    return ''.join(random.choices(string.ascii_letters + string.digits, k=length)) + '\n'

# File sizes in GB
file_sizes_gb = [1, 2, 3, 4, 5]

for size_gb in file_sizes_gb:
    target_size = size_gb * 1024 * 1024 * 1024  # Convert GB to bytes
    filename = f"random_strings_{size_gb}GB.txt"

    print(f"Creating {filename} ({size_gb} GB)...")
    with open(filename, "w") as f:
        while f.tell() < target_size:
            f.write(random_string())

    print(f"{filename} created: {os.path.getsize(filename)} bytes")
```

```python
# ques 6
import os
import threading

def convert_file_to_upper(input_file, output_file):
    print(f"[START] Converting {input_file} -> {output_file}")
    with open(input_file, "r") as infile, open(output_file, "w") as outfile:
        for line in infile:
            outfile.write(line.upper())
    print(f"[DONE]  {output_file} created ({os.path.getsize(output_file)} bytes)")

# File sizes in GB
file_sizes_gb = [1, 2, 3, 4, 5]
threads = []

# Start a thread for each file
for size_gb in file_sizes_gb:
    input_filename = f"random_strings_{size_gb}GB.txt"
    output_filename = f"random_strings_{size_gb}GB_UPPER.txt"

    thread = threading.Thread(target=convert_file_to_upper, args=(input_filename, output_filename))
    thread.start()
    threads.append(thread)

# Wait for all threads to finish
for thread in threads:
    thread.join()

print("âœ… All files have been converted to uppercase in parallel.")
```

```python
#QUES 12
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#(i) Create a random dataset of 100 rows and 30 columns ---
print("--- Step 1: Creating a random dataset (100 rows, 30 columns) ---")
# Generate random integers between 1 and 200
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data, columns=[f'col_{i}' for i in range(30)])
print("Original DataFrame head:")
print(df.head())
print("\n" + "="*80 + "\n")

# (ii) Replace all the values with NA in the dataset defined between [10, 60] ---
print("--- Step 2: Replacing values between [10, 60] with NA ---")
# Create a boolean mask where values are between 10 and 60 (inclusive)
mask = (df >= 10) & (df <= 60)
df_with_na = df.where(~mask) # Replace values where mask is True with NaN (NA)

print("DataFrame head after replacing values with NA:")
print(df_with_na.head())

# Print the count of number rows having missing values
rows_with_missing_values = df_with_na.isnull().any(axis=1).sum()
print(f"\nNumber of rows with missing values (NA): {rows_with_missing_values}")
print("\n" + "="*80 + "\n")

#(iii) Replace all the NA values with the average of the column value ---
print("--- Step 3: Replacing NA values with the average of their respective columns ---")
df_filled = df_with_na.fillna(df_with_na.mean())
print("DataFrame head after filling NA values with column averages:")
print(df_filled.head())
print("\n" + "="*80 + "\n")

#(iv) Find the Pearson correlation among all the columns and plot heat map ---
print("--- Step 4: Pearson Correlation Heatmap & Column Selection ---")
# Calculate Pearson correlation matrix
correlation_matrix = df_filled.corr(method='pearson')

# Plotting the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Pearson Correlation Heatmap of Dataset Columns')
plt.show()

# Select those columns having correlation <= 0.7
# We will identify pairs of columns where the absolute correlation coefficient is <= 0.7
print("\nPairs of columns with absolute Pearson correlation <= 0.7:")
low_correlation_pairs = []
# Iterate through the upper triangle of the correlation matrix to avoid duplicates
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        col1 = correlation_matrix.columns[i]
        col2 = correlation_matrix.columns[j]
        corr_value = correlation_matrix.loc[col1, col2]
        if abs(corr_value) <= 0.7:
            low_correlation_pairs.append((col1, col2, corr_value))

if low_correlation_pairs:
    for pair in low_correlation_pairs:
        print(f"  {pair[0]} - {pair[1]}: Correlation = {pair[2]:.4f}")
else:
    print("  No pairs of columns found with absolute correlation <= 0.7.")
print("\n" + "="*80 + "\n")

#(v) Normalize all the values in the dataset between 0 and 10 ---
print("--- Step 5: Normalizing all values in the dataset between 0 and 10 ---")
df_normalized = df_filled.copy() # Work on a copy to preserve previous states

# Min-Max Scaling formula: X_scaled = (X - X_min) / (X_max - X_min) * (new_max - new_min) + new_min
# Here, new_min = 0, new_max = 10
# We apply this column-wise to ensure correct min/max for each column
for column in df_normalized.columns:
    min_val = df_normalized[column].min()
    max_val = df_normalized[column].max()

    # Avoid division by zero if all values in a column are the same
    if (max_val - min_val) != 0:
        df_normalized[column] = ((df_normalized[column] - min_val) / (max_val - min_val)) * 10
    else:
        # If all values are the same, they should all map to the new_min (0 in this case)
        df_normalized[column] = 0

print("DataFrame head after normalization (values between 0 and 10):")
print(df_normalized.head())
print("\n" + "="*80 + "\n")

#(vi) Replace all the values in the dataset with 1 if value <= 0.5 else with 0 ---
print("--- Step 6: Replacing values: 1 if <= 0.5, else 0 ---")
# Apply a lambda function across the entire DataFrame
df_binary = df_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)

print("DataFrame head after binary transformation:")
print(df_binary.head())
print("\n" + "="*80 + "\n")

# Optional: Display some statistics of the final binary dataset
print("Value counts for a sample column (e.g., col_0) after binary transformation:")
print(df_binary['col_0'].value_counts())
```

```python
#QUES 14
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Generate a random dataset ---
# Set seed for reproducibility
np.random.seed(42)

# Generate 600 rows and 15 columns with values between -100 and 100
num_rows = 600
num_cols = 15
data = np.random.uniform(low=-100, high=100, size=(num_rows, num_cols))

# Create a Pandas DataFrame for easier handling
column_names = [f'Column {i+1}' for i in range(num_cols)]
df = pd.DataFrame(data, columns=column_names)

print("Generated Dataset Head:")
print(df.head())
print("\nDataset Info:")
df.info()
print("\nDataset Description:")
print(df.describe())

# --- 2. Plot scatter graph between Column 5 and Column 6 ---
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Column 5', y='Column 6', data=df, alpha=0.6)
plt.title('Scatter Plot: Column 5 vs Column 6')
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --- 3. Plot histogram of each column in single graph ---
# Determine grid size for subplots
rows = int(np.ceil(np.sqrt(num_cols)))
cols = int(np.ceil(num_cols / rows))

plt.figure(figsize=(4 * cols, 3 * rows))
plt.suptitle('Histograms of All Columns', y=1.02, fontsize=16) # Add a main title

for i, col in enumerate(df.columns):
    plt.subplot(rows, cols, i + 1) # i+1 because subplot is 1-indexed
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f'{col} Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.tight_layout() # Adjust layout to prevent overlapping titles/labels

plt.show()

# --- 4. Plot the Box plot of each column in single graph ---
plt.figure(figsize=(4 * cols, 3 * rows))
plt.suptitle('Box Plots of All Columns', y=1.02, fontsize=16) # Add a main title

for i, col in enumerate(df.columns):
    plt.subplot(rows, cols, i + 1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'{col} Box Plot')
    plt.ylabel('Value')
    plt.tight_layout() # Adjust layout to prevent overlapping titles/labels

plt.show()
```
